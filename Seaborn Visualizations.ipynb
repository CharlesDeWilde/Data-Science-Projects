```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# setting and checking color palette: deep, muted, pastel, bright, dark, and colorblind
sns.set_palette("deep")
sns.palplot(sns.color_palette())
```

```python
# load data
# https://github.com/mwaskom/seaborn-data
normd = np.random.normal(size=100)
gammad = np.random.gamma(6, size=100)
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
##### For reference:
#distributions
# https://seaborn.pydata.org/tutorial/distributions.html
#  - sns.distplot() - figure-leve interface for univariate distributions (encompases histogram, kdeplot, rugplot, and fitting a parametric distribution to data)
#  - sns.jointplot() - figure-level interface for bivariate distributions (plt.scatter, plt.hexbin, kde)
#  - sns.pairplot() - figure-level interface to plot multiple pairwise bivariate distributions in a dataset
```

```python
# univariate: histogram
sns.distplot(normd,hist=True,kde=False,rug=False,bins=20)
```

```python
# univariate: rug plot
# draws a small vertical tick at each observation
sns.distplot(normd, hist=False, kde=False, rug=True)
```

```python
# univariate: kde (kernel density estimate) plot
# - The kernel density estimate can be a useful tool for plotting the shape of a distribution.
# - Like the histogram, the KDE plots encode the density of observations on one axis with height along the other axis.
# - Drawing a KDE is more computationally involved than drawing a histogram.
# - Each observation is first replaced with a normal (Gaussian) curve centered at that value.
# - These curves are then summed to compute the value of the density at each point in the support grid.
# - The resulting curve is then normalized so that the area under it is equal to 1.
sns.distplot(normd, hist=False, kde=True, rug=False) 
```

```python
sns.distplot(normd,hist=True,kde=True,rug=True,bins=20)
```

```python
# univariate: fit a parameteric distribution to a dataset to visually evaluate how closely it corresponds to the observed data
from scipy import stats
sns.distplot(normd,hist=True,kde=False,rug=False,bins=20, fit= stats.gamma)
```

```python
# bivariate: scatter plot
# jointplot (kind=scatter)
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

```python
# bivariate: hexbin (shows the counts of observations that fall within hexagonal bins)
# jointplot (kind=hex)
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex', color='r')
```

```python
# bivariate: kde
# jointplot (kind=kde)
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

```python
# bivariate: reg (linear relationship between variables in a dataset)
# jointplot (kind=reg)\n",
sns.jointplot(tips, x='total_bill', y='tip', kind='reg',color='g')
```

```python
# pairplot: to plot multiple pairwise bivariate distributions in a dataset
sns.pairplot(iris, kind='scatter', diag_kind='hist')
```

```python
sns.pairplot(iris, kind='reg', diag_kind='hist')
```

```python
sns.pairplot(iris, kind='scatter', diag_kind='kde')
```

```python
sns.pairplot(iris, kind='scatter', diag_kind='hist', hue= 'species')
```

```python
sns.pairplot(tips, kind='scatter', diag_kind='hist', hue= 'sex')
```
